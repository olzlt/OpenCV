# cv.2getRotationMatrix2D 

OpenCV  cv.2getRotationMatrix2D  

' '  

cv2.getRotationMatrix2D(center, angle, scale) 

center -   . (x, y)

angle -     (degree)

scale -  

 Affine   



```python
# 90 

cp = (img.shape[1] / 2, img.shape[0] / 2) #   1/2,  1/2
rot = cv2.getRotationMatrix2D(cp, -90, 1) # -90 ( 90),  1

dst = cv2.warpAffine(img, rot, (0, 0))

cv2_imshow(img)
cv2_imshow(dst)
```

```python
# 45 

cp = (img.shape[1] / 2, img.shape[0] / 2) #   1/2,  1/2
rot = cv2.getRotationMatrix2D(cp, 45, 0.7) # -90 ( 90),  0.7

dst = cv2.warpAffine(img, rot, (0, 0))

cv2_imshow(img)
cv2_imshow(dst)
```
